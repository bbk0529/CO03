```python
import pandas as pd
```

```python
BOM={}
with open("CO03.txt") as files :
    for filename in files : 
        lines=open(filename.replace("\n","")).readlines()
        for line in lines[10:-1] :    
            spltline=line.split("|")    
            partnr=spltline[2].strip()
            qty=int(float(spltline[4].strip()))
            BOM[partnr] = BOM.get(partnr, 0) + qty    
```

```python
pd.DataFrame(data=[BOM.keys(), BOM.values()]).transpose()
```
